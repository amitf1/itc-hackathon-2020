```python
import pandas as pd
```

```python
pd.set_option('display.max_columns', None)  
```

```python
data1_path = "dataset.csv"
data2_path = "dataset2.csv"
```

```python
df = pd.read_csv(data1_path)
df.head()
```

```python
df2 = pd.read_csv(data2_path)
df2.head()
```

Combine the important columns of datasets:

```python
sums = pd.concat([df["summary"], df2["Summary"]], ignore_index=True)
titles = pd.concat([df["title"], df2["opportunity_id"]], ignore_index=True)
categories = pd.concat([df["category_desc"], df2["Category Description"]], ignore_index=True)
```

```python
joined = pd.DataFrame()
joined["sum"] = sums
joined["title"] = titles
joined["category"] = categories
```

```python
joined.to_csv
```
